```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## **DATA DESCRIPTION**
PRODUCT - (A - F)<br/>

AGE - age of the borrower<br/>

AREA - residential area ('County capital' 'Urban area' 'Rural area' 'Missing')<br/>

RESIDENTIAL PLACE - housing situation ('Owner without mortgage' 'Living with family' 'Owner with mortgage'
'Rental' 'Other')<br/>

EDUCATION - education level ('University' 'Post-graduate' 'High school' 'Missing'
'Post-secondary school' 'Other' 'College' 'Vocational school'
'Middle school' 'Primary school')<br/>

MARITAL STATUS - marital status ('married' 'divorced' 'single' 'widow')<br/>

HOUSEHOLD MEMBERS - number of people in the household<br/>

NO OF DEPENDENTS - number of dependents of the client<br/>

INCOME - income ($/month ?)<br/>

WORK SENIORITY - work experience<br/>

BUSINESS AGE - business experience (if self-employed)<br/>

ECONOMIC SECTOR - sector in which the borrower works<br/>

EMPLOYEE NO - number of employees<br/>

LENGTH RELATIONSHIP WITH CLIENT - duration of the business relationship with the client<br/>

DEBIT CARD - whether the client has a debit card<br/>

CURRENT ACCOUNT - whether the client has a current account<br/>

SAVING ACCOUNT - whether the client has a savings account<br/>

SALARY ACCOUNT - whether the client has a salary account<br/>

FOREIGN ACCOUNT - whether the client has a foreign currency account<br/>

FINALIZED LOAN - whether the client has repaid the loan<br/>

DEPOSIT - whether the client has a deposit<br/>

PENSION FUNDS - whether the client has pension funds<br/>

**DEFAULT FLAG** - whether the client has payment arrears (**1 - yes, 0 - no**)<br/>

## **Outliers**

### LENGTH_RELATIONSHIP_WITH_CLIENT

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (18, 8))

sns.violinplot(x = df["LENGTH_RELATIONSHIP_WITH_CLIENT"], ax = axes[0, 0])
sns.boxplot(x = df["LENGTH_RELATIONSHIP_WITH_CLIENT"], ax = axes[0, 1])
sns.histplot(x = df["LENGTH_RELATIONSHIP_WITH_CLIENT"], ax = axes[1, 0], kde = True, stat = "density", linewidth = 0)
sns.histplot(x = np.log1p(df["LENGTH_RELATIONSHIP_WITH_CLIENT"]), ax = axes[1, 1], kde=True, stat = "density", linewidth = 0)

fig.suptitle("LENGTH_RELATIONSHIP_WITH_CLIENT - DISTRIBUTION", size = 20)
axes[1, 0].set_xlabel("LENGTH_RELATIONSHIP_WITH_CLIENT")
axes[1, 1].set_xlabel("log(LENGTH_RELATIONSHIP_WITH_CLIENT)")

plt.show()
```

```python
z = np.abs(stats.zscore(df["LENGTH_RELATIONSHIP_WITH_CLIENT"]))
outliers = df[z > 3]
print("number of outliers: " + str(len(outliers)))
print("% of outliers: " +  str(round(len(outliers) * 100 / len(df), 3)) + " %")
```

```python
import matplotlib.pyplot as plt

residential_places = df["RESIDENTIAL_PLACE"].unique()
fig, axes = plt.subplots(nrows=1, ncols=len(residential_places), figsize=(6 * len(residential_places), 4))

for i, place in enumerate(residential_places):
    # Filtrowanie danych dla danego miejsca zamieszkania
    subset_df = df[df["RESIDENTIAL_PLACE"] == place]
    
    # Obliczanie wartoÅ›ci dla DEFAULT_FLAG == 0
    flag_0 = len(subset_df[subset_df["DEFAULT_FLAG"] == 0])
    
    # Obliczanie wartoÅ›ci dla DEFAULT_FLAG == 1
    flag_1 = len(subset_df[subset_df["DEFAULT_FLAG"] == 1])
    
    # Obliczanie procentowej wartoÅ›ci dla DEFAULT_FLAG == 0
    percentage_0 = flag_0 * 100 / (flag_0 + flag_1)
    
    # Obliczanie procentowej wartoÅ›ci dla DEFAULT_FLAG == 1
    percentage_1 = flag_1 * 100 / (flag_0 + flag_1)
    
    # Rysowanie skumulowanego sÅ‚upka
    axes[i].bar(0, percentage_0, label="0")
    axes[i].bar(0, percentage_1, bottom=percentage_0, label="1")
    
    # Dodawanie etykiet i tytuÅ‚Ã³w
    axes[i].set_ylabel("%")
    axes[i].set_title(place)
    axes[i].legend()

# Dostosowanie ukÅ‚adu
plt.tight_layout()

# WyÅ›wietlenie wykresu
plt.show()
```
